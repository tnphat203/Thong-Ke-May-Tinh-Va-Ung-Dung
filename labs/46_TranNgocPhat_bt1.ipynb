1. Viết chương trình nhập vào số nguyên n (n>0). In ra các số nguyên tố từ 1..n


```python
def isSoNguyenTo(num):
    if(num<=1):
        return False
    for i in range(2,int(num**0.5)+1):
        if num % i == 0:
            return False
    return True

def inSoNguyenTo(num):
    danhSach = []
    for i in range(2,num+1):
        if(isSoNguyenTo(i)):
            danhSach.append(i)
    print("Danh sách số nguyên tố từ 0 đến n: ", end = "")
    print(danhSach)

n = int(input("Nhập n (n>0): "))
inSoNguyenTo(n)
```

2. Viết chương trình nhập vào số nguyên n có 3 chữ số. Tính tổng các chữ số của n

```python
n = int(input("Nhập n: "))
if n>=100 | n<=999:
    print("Phải nhập số nguyên có 3 chữ số")
    exit(1)
else:
    sum = 0
    while n!=0:
        sum+=n%10
        n=int(n/10)
    print("Tổng các chữ số của n: ",end = "")
    print(sum)
```

3. Viết chương trình nhập vào một dãy các số nguyên được cách nhau bởi khoảng trắng (VD: 32 8 4 6 9...) lưu vào python ở dạng mảng. Thực hiện

- Xuất mảng
- Tính tổng các phần tử của mảng
- Tìm số lớn nhất
- Tìm số bé nhất
- Sắp xếp mảng theo thứ tự giảm dần
- In ra các số chẵn trong mảng

```python
danhSach = input().split()
danhSach = [int(i) for i in danhSach]
danhSach2 = []
sum = 0
print("Danh sách số nguyên: ", end = "")
print(danhSach)
for i in danhSach:
    sum+=i
print("Tổng các phần tử của mảng: ", end = "")
print(sum)
print("Số lớn nhất: ", end = "")
print(max(danhSach))
print("Số bé nhất: ", end = "")
print(min(danhSach))
print("Danh sách sau khi sắp xếp: ", end = "")
print(sorted(danhSach, reverse=True))
print("Danh sách các số chẵn: ", end = "")
for i in danhSach:
    if i % 2 == 0:
       danhSach2.append(i)
print(danhSach2)
```

4. Viết chương trình yêu cầu người dùng nhập họ tên. Hiển thị tên người dùng sau khi đã chuẩn hóa. Biết rằng tên đã chuẩn hóa là:
- Ký tự đầu viết hoa
- Không có khoảng trắng thừa
- Không có ký tự đặc biệt

```python
def chuanHoaHoTen(hoTen):
    hoTenThuong = hoTen.lower()
    tachHoTen = hoTenThuong.split()
    chuanHoaTu = [t.capitalize() for t in tachHoTen]
    hoTenChuanHoa = ' '.join(chuanHoaTu)
    return hoTenChuanHoa
    
hoTen = input("Nhập vào họ tên: ")
hoTenCH = chuanHoaHoTen(hoTen)
print("Họ tên sau khi chuẩn hóa: ", end="")
print(hoTenCH)
```
