```python
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm

import matplotlib.pyplot as plt
import numpy as np
```

**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
#a 
n, p = 10, 0.25
print('Số câu đúng của 10 câu hỏi là: ', binom.rvs(n, p))
```

```python
#b
#X là số câu đúng
#Y là điểm số đạt được
#Y = 4X - 2(10 - X) = 4 => Số câu đúng là X = 4
print('Xác suất đạt được 4 điểm là: ', binom.pmf(4, n, p))
```

```python
#c
E_X = n * p
Var_X = n * p * (1 - p)
print('Kì vọng số câu đúng: ', E_X)
print('Phương sai số câu đúng: ', Var_X)
```

**Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
lamda = 3
print('Xác suất để nhận được 1 cuộc gọi: ', poisson.pmf(1, lamda))
print('Xác suất để nhận được 1 cuộc gọi: ', poisson.pmf(2, lamda))
print('Xác suất để nhận được 1 cuộc gọi: ', poisson.pmf(3, lamda))
```

Loại 1: trên 505 gam

Loại 2: từ 495 – 505 gam

Loại 3: dười 495 gam

a. Hãy tính tỷ lệ của mỗi loại.

b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

```python
#sigma = sqrt(16)
X = norm(500, 4)
#a Tính tỷ lệ mỗi loại
Loai1 = 1 - X.cdf(505)
Loai2 = Loai1 - X.cdf(495)
Loai3 = X.cdf(495)
print('Tỷ lệ loại 1: ', Loai1)
print('Tỷ lệ loại 2: ', Loai2)
print('Tỷ lệ loại 3: ', Loai3)
```

```python
#b
mau = X.rvs(20)
plt.hist(mau)
plt.title('Biểu đồ cho 20 mẫu')
plt.xlabel('Số gram')
plt.ylabel('Tần suất')
plt.show()
print('Giá trị trung bình: ', mau.mean())
print('Độ lệch chuẩn: ', mau.std())
```

```python
#c
mau = X.rvs(100)
plt.hist(mau)
plt.title('Biểu đồ cho 100 mẫu')
plt.xlabel('Số gram')
plt.ylabel('Tần suất')
plt.show()
print('Giá trị trung bình: ', mau.mean())
print('Độ lệch chuẩn: ', mau.std())
```

**Bài 04:** Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.

```python
#a
size = 5000
mat4 = 0
for _ in range(size):
    if np.random.randint(1, 7) == 4:
        mat4 += 1
print('Xác suất để tung ra mặt 4: ', mat4 / size)
```

```python
#b
size = 5000
mat456 = 0
for _ in range(size):
    if np.random.randint(1, 7) >= 4:
        mat456 += 1
print('Xác suất để tung ra mặt lớn hơn hoặc bằng 4: ', mat456 / size)
```

```python
#b
size = 5000
mat456 = 0
mat6 = 0
for _ in range(size):
    n = np.random.randint(1, 7)
    if n >= 4:
        mat456 += 1
    if n == 6:
        mat6 += 1
print('Xác suất để mặt 6 chấm xuất hiện: ', mat6 / mat456)
```

**Bài 05:** Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7 Bạn Hãy

a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3

b. Vẽ đồ thị boxplot

c. Vẽ các đồ thị: tần số, tần suất

```python
X = binom(50, 0.7)
mau = X.rvs(5000)
```

```python
#a
print('Min: ', mau.min())
print('Max: ', mau.max())
print('Q1: ', np.quantile(mau, 0.25))
print('Q2: ', np.quantile(mau, 0.5))
print('Q3: ', np.quantile(mau, 0.75))
```

```python
#b
plt.boxplot(mau)
plt.show()
```

```python
#c
plt.hist(mau)
plt.show()
```
