## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

````python
import numpy as np
x = np.random.randint(1000,size=100)
print("10 phần tử đầu tiên:",x[:10])
print("5 phần tử cuối cùng:",x[-5:])
print("Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100:",x[(x%5==0)&(x>100)])
````

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

````python
print("Các phần tử lớn hơn 100:",x[x>100])
````

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

````python
print("Tính tổng mảng:",np.sum(x))
print("Tính giá trị trung bình:",np.mean(x))
print("Tính giá trị trung vị:",np.median(x))
print("Q1 là:",np.percentile(x,25))
print("Q2 là:",np.percentile(x,50))
print("Q3 là:",np.percentile(x,75))
print("Phân vị 20 là:",np.percentile(x,20))
````

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

````python
for i in range(5):
    random_selection = np.random.choice(x, size=10, replace=False)
    print(f"Random selection {i+1}: {random_selection}")
````

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

````python

a = np.bincount(x).argmax()
print("The most frequently occurring element is:", a)
````
